# 通道，成交量繪圖

````python
from matplotlib import pyplot as plt
from matplotlib import style

style.use('ggplot')
 

data = df

price = data['Close']
price.head()
 
moving_avg = price.rolling(20).mean()
moving_avg_mstd = price.rolling(20).std()

# 新增圖像 用以存取
fig = plt.figure(figsize=(10,10))
# 分格 2 index 1 col 第1張
top = plt.subplot(211)
# 分格 2 index 1 col 第2張
bottom = plt.subplot(212)
 
top.plot(price, color='b')
top.plot(moving_avg, color='r')

top.fill_between(
    moving_avg.index, 
    moving_avg-2*moving_avg_mstd, 
    moving_avg+2*moving_avg_mstd, 
    color='b', alpha=0.2
)
bottom.bar(data.index, data['Volume'])
 
plt.show()

````

````python
from io import BytesIO
import base64


# 写入内存
save_file = BytesIO()
fig.savefig(save_file, format='jpg')
fig.savefig('111.jpg', format='jpg')
# 转换base64并以utf8格式输出
save_file_base64 = base64.b64encode(save_file.getvalue())#.decode('utf8')
save_file_base64
````
