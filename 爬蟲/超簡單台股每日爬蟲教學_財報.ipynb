# 連續 N 天

```python
import requests
from io import StringIO
import pandas as pd
import numpy as np
```

```python
def scrap(datestr):
    '''
    datestr = '20180131'
    '''
    
    r = requests.post(
        'http://www.twse.com.tw/exchangeReport/MI_INDEX?response=csv&date='
        + datestr + '&type=ALL'
    )

    df = pd.read_csv(
        StringIO(
            "".join(
                [
                    i.translate({ord(c): None for c in ' '}) 
                        for i in r.text.split('\n') 
                        if len(i.split('",')) == 17 and i[0] != '='
                ]
            )
        ), header=0, index_col=[0]
    ).iloc[:,:-1]
    
    df = df.loc[:,['證券名稱','成交股數','開盤價','最高價','最低價','收盤價','本益比']]
    
    df['日期'] = pd.to_datetime(datestr,format='%Y/%m/%d')
    
    return df
```

```python
from datetime import (
    timedelta, date , datetime
)
import time
```

```python
start = '2018/1/1'
end = '2018/3/31'

start_day = datetime.strptime(start,'%Y/%m/%d')
end_day = datetime.strptime(end,'%Y/%m/%d')

day = start_day
# to string
#day.strftime('%Y/%m/%d')

interval = timedelta(days=1)

# 時間間距
span = (end_day - start_day).days + 1

day_df_list = []
eorr_day = []
for i in range(span):
    try:
        day_df_list.append(
            scrap(day.strftime('%Y/%m/%d'))
        )
        print(day.strftime('%Y/%m/%d'))
        
    except Exception as e:
        eorr_day.append(day.strftime('%Y/%m/%d'))
        print(f"The day have eorr: {day.strftime('%Y/%m/%d')}" + "---"*10)
    
    finally:
        day += interval 
        time.sleep(15)
    
```

```python
df = pd.concat(day_df_list)
```

# Python 財報爬蟲

https://www.finlab.tw/Python-%E8%B2%A1%E5%A0%B1%E7%88%AC%E8%9F%B2-1-%E7%B6%9C%E5%90%88%E6%90%8D%E7%9B%8A%E8%A1%A8/

```python
import requests
import pandas as pd
import numpy as np

def financial_statement(year, season, type='綜合損益彙總表'):

    
    year -= 1911 # 西元轉民國
        
    if type == '綜合損益彙總表':
        url = 'http://mops.twse.com.tw/mops/web/ajax_t163sb04'
    elif type == '資產負債彙總表':
        url = 'http://mops.twse.com.tw/mops/web/ajax_t163sb05'
    elif type == '營益分析彙總表':
        url = 'http://mops.twse.com.tw/mops/web/ajax_t163sb06'
    else:
        print('type does not match')

    r = requests.post(url, {
        'encodeURIComponent':1,
        'step':1,
        'firstin':1,
        'off':1,
        'TYPEK':'sii',
        'year':str(year),
        'season':str(season),
    })
    
    r.encoding = 'utf8'
    dfs = pd.read_html(r.text)
    
    
    for i, df in enumerate(dfs):
        df.columns = df.iloc[0]
        dfs[i] = df.iloc[1:]
        
    df = pd.concat(dfs).applymap(lambda x: x if x != '--' else np.nan)
    df = df[df['公司代號'] != '公司代號']
    df = df[~df['公司代號'].isnull()]
    return df
```

```python
financial_statement(2017,1,'營益分析彙總表')
```

```python
financial_statement(2017,1,'資產負債彙總表')
```
