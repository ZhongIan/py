```python
import json
jd = json.loads(soup.select_one('#local-page-info').text)
jd
```

```python
import pandas as pd
products = []
for k in jd['data']:
    if 'list' in jd['data'][k]:
        df  = pd.DataFrame(jd['data'][k]['list'])
        products.append(df)
```

```python
products_all = pd.concat(products)
products_all
```

```python
products_all.columns
```

```python
products_all = products_all.loc[
    products_all['itemTitle'].notnull(), 
    ['itemUrl', 'itemTitle', 'itemDesc','itemMonthSoldCount','itemActPrice', 'itemTagPrice']
]
```

```python
products_all.info()
```

```python
products_all['itemActPrice'] = products_all['itemActPrice'].astype(float)
products_all['itemTagPrice'] = products_all['itemTagPrice'].astype(float, inplace=True)
```

```python
products_all['discount'] = products_all['itemTagPrice'] - products_all['itemActPrice']
```

```python
products_all.sort_values('discount', ascending=False)
```
